Matplotlib to biblioteka w Pythonie służąca do tworzenia statycznych, animowanych i interaktywnych wizualizacji w Pythonie. Matplotlib może generować wykresy, histogramy, widoki mocy, wykresy błędów, wykresy punktowe itp., za pomocą tylko kilku linii kodu.

Aby zacząć korzystać z Matplotlib, musimy najpierw zaimportować bibliotekę. Najczęściej importujemy pod nazwą 'plt' dla wygody.

```python
import matplotlib.pyplot as plt
```

## Tworzenie prostych wykresów liniowych
Najprostszym wykresem, który możemy stworzyć, jest wykres liniowy. Wykres liniowy łączy punkty danych w porządku, w którym są one podane. Możemy stworzyć wykres liniowy, podając listę liczb jako argument do funkcji `plt.plot()`. Po stworzeniu wykresu, możemy go wyświetlić za pomocą funkcji `plt.show()`.

```python
plt.plot([1, 2], [3,8])
plt.show()
```

## Dodawanie tytułów, etykiet osi i legend do wykresów
Możemy dodać tytuł do naszego wykresu za pomocą funkcji `plt.title()`, a etykiety osi za pomocą funkcji `plt.xlabel()` i `plt.ylabel()`. Możemy również dodać legendę do wykresu za pomocą funkcji `plt.legend()`. W przypadku wykresu liniowego, musimy dodać argument `label` do funkcji `plt.plot()`, aby określić, co ma być wyświetlane w legendzie.

```python
plt.plot([1, 2, 3, 4, 5], linestyle='--', label='Dollar Price')
plt.plot([2, 4, 6, 8, 10], color='red', label='PLN Price')
plt.plot([5, 4, 3, 2, 1], linestyle=':', label='DK Price')
plt.title('Chart Title')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

## Zmiana stylów i kolorów na wykresach
Możemy zmienić styl i kolor naszego wykresu, dodając więcej argumentów do funkcji `plt.plot()`. Możemy określić kolor linii, dodając argument `color`, a styl linii, dodając argument `linestyle`. Możemy również zmienić grubość linii, dodając argument `linewidth`.

```python
plt.plot([1, 2, 9, 3, 5], label='Niebieska linijka', color='blue', linestyle='-.', linewidth=3)
plt.title('Nasz Drugi Wykres')
plt.xlabel('X Oś')
plt.ylabel('Y Oś')
plt.legend()
plt.show()
```
